#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
# TODO
# Normalizar el precio para cada población
subconjunto['precio_normalizado'] = subconjunto.groupby('level5')['price'].transform(lambda x: (x - x.min()) / (x.max() - x.min()))

plt.figure(figsize=(10, 6))

for poblacion in ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']:
    datos = subconjunto[subconjunto['level5'] == poblacion]['precio_normalizado']
    plt.hist(datos, bins=20, alpha=0.5, label=poblacion)

# Configuración de etiquetas y título
plt.title('Histogramas de Precios Normalizados por Población', fontsize=16)
plt.xlabel('Precio Normalizado', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.legend(title='Poblaciones')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆) 
Pista: Haz un diagrama de dispersión de cada población usando subgráficos (subplots).

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt

poblaciones_deseadas = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']
subconjunto = subconjunto[subconjunto['level5'].isin(poblaciones_deseadas)]

subconjunto['pps'] = subconjunto['price'] / subconjunto['surface']

fig, axs = plt.subplots(2, 2, figsize=(12, 10))  # 2 filas, 2 columnas
fig.suptitle('Precio por m² vs Superficie en Diferentes Poblaciones', fontsize=16)

for ax, poblacion in zip(axs.flatten(), poblaciones_deseadas):
    data_pob = subconjunto[subconjunto['level5'] == poblacion]
    ax.scatter(data_pob['surface'], data_pob['pps'], alpha=0.7, c='steelblue', edgecolor='k')
    ax.set_title(poblacion)
    ax.set_xlabel('Superficie (m²)')
    ax.set_ylabel('Precio por m²')
    ax.grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
